### 1. Importando Bibliotecas

```python
import pandas as pd
import numpy as np
```

### 2. Importando e Limpando a Base de Dados

```python
df = pd.read_csv(r"C:\Users\Joaquim Bermudes\Downloads\producao_pecuaria\producao_pecuaria.csv")
df.head()
```

```python
n_municipios = len(pd.unique(df["id_municipio"])) #descobrindo quantos municípios tem na base
vacas = np.array(df["vacas_ordenhadas"]).reshape((46, n_municipios)) #separando a coluna "vacas_ordenhadas" por município
galinhas = np.array(df["ovinos_tosquiados"]).reshape((46, n_municipios)) #separando a coluna "ovinos_tosquiados" por município
```

```python
Vacas = pd.DataFrame(vacas) #transformando o array de "vacas" em um dataframe
Galinhas = pd.DataFrame(galinhas) #tranformando o array de "galinhas" em um dataframe
```

```python
Vacas.index = pd.date_range("1974", "2020", freq = "Y").year #colocando os anos nos índices
Galinhas.index = pd.date_range("1974", "2020", freq = "Y").year #colocando os anos nos índices
```

```python
Vacas.columns = pd.unique(df["id_municipio"]) #colocando o "id_municipio" como nome das séries
Galinhas.columns = pd.unique(df["id_municipio"]) #colocando o "id_municipio" como nome das séries
```

```python
#mantendo as somente séries de que tenham os ultimas 30 observações em "Vacas" 
Continua = [True]*len(Vacas.columns)
for i in range(len(Vacas.columns)):
    if (Vacas.iloc[15:46, i].isna().sum() != 0):
        Continua[i] = False

Vacas = Vacas.iloc[:, Continua]

#mantendo as somente séries de que tenham os ultimas 30 observações em "Galinhas" 
Continua = [True]*len(Galinhas.columns)
for i in range(len(Galinhas.columns)):
    if (Galinhas.iloc[15:46, i].isna().sum() != 0):
        Continua[i] = False

Galinhas = Galinhas.iloc[:, Continua]
```
